```python
import pandas as pd
import numpy as np
import datetime
```

Import billboard csv and display the top 10.

```python
billboard = pd.read_csv('https://raw.githubusercontent.com/staceyvo/IS362_Project2/master/billboard_data.csv')
billboard.head(10)
```

We will melt this table to make the data easier to read.

```python
id_vars = ['year','artist.inverted','track','time','genre','date.entered','date.peaked']
billboard = pd.melt(frame=billboard,id_vars=id_vars, var_name='week', value_name='rank')
billboard.head(10)
```

We will fomat the data, remove NaN values, and we will remove the extra data from the week column, leaving behind only an integer.

```python
billboard['week'] = billboard['week'].astype(str)
billboard = billboard.dropna()
billboard['week'] = billboard['week'].str.extract('(\d+)', expand=False).astype(int)
billboard['rank'] = billboard['rank'].astype(int)
billboard.head(10)
```

Next, we will group the data to see what genres are represented in this dataset.

```python
billboard.groupby('genre').genre.count()
```

Next we will create a table of songs, removing duplicate listings.

```python
songs_cols = ['year', 'artist.inverted', 'track', 'time', 'genre']
songs = billboard[songs_cols].drop_duplicates()
songs = songs.reset_index(drop=True)
songs["song_id"] = songs.index
songs.head(10)
```

Now we will take a look at 311 Service requests.

Let's start by importing the 311 Service Requests from January in order to take a look at the data.

```python
serv_req = pd.read_csv('https://raw.githubusercontent.com/staceyvo/IS362_Project2/master/311_Service_Requests_2011_Jan.csv')
serv_req
```

Let's rename the columns to make the data easier to work with.

```python
serv_req.columns = [c.replace(' ', '_') for c in serv_req.columns]
serv_req.head(10)
```

First, we will start by looking at how many service requests came from each borough.

```python
by_borough = serv_req.groupby('Borough').Complaint_Type.count().reset_index(name='Count')
by_borough
```

What did people complain about most during the month of January? Snow

```python
complaint_type = serv_req.groupby('Complaint_Type').Complaint_Type.count().reset_index(name='Count')
complaint_type
```

Importing Titanic data

```python
titanic_data = pd.read_csv('https://raw.githubusercontent.com/staceyvo/IS362_Project2/master/titanic_data.csv')
titanic_data.head(10)
```

For age, let's replace NaN with 0 and caste type as integer

```python
titanic_data['age'] = titanic_data.age.fillna(0).astype(int)
titanic_data.head(10)
```

We have a few columns that are missing a lot values and we don't need them for analysis, so we will remove them.

```python
titanic = titanic_data.drop(titanic_data.columns[[5,6, 12]], axis=1)
titanic.head(10)
```

Now we'll see the survivors broken down by class.

```python
passengers = titanic.groupby('pclass').name.count()
passengers
```

Now that we know how many passengers there were by class, let's see how many survivors there were by class.

```python
survivors_class = titanic.groupby('pclass').survived.sum()
survivors_class
```

What was the survivor breakdown by gender?

```python
survivors_sex = titanic.groupby('sex').survived.sum()
survivors_sex
```

What was the average fare paid by gender?

```python
fare = titanic.groupby('sex').fare.mean()
fare
```
